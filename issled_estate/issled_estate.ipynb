**Выделение типов ошибок в data:**


Случайные ошибки:
Значения rooms 1 или 0 и Значения False в studio при указании квартиры как студии или после перепланировки. Здесь ошибка возникает из-за сравнительно недавнего появления нового типа квартир - студий. Многие продавцы склонны считать студии или квартиры переделанные под студии как безкомнатные.  

Значения NaN в столбце kitchen_area. Возможный сбой при загрузке данных. Легко восстановить имея данные об общей площади из-за заметной корреляции.
Значения NaN в столбце balcony. Либо сбой, либо отсутсвие балконов в квартире.
Пропуски в ceiling_height, floors_total - никто с линейкой не меряет, кадастровый план мало кто склонен подробно изучать. Этажность больше 10-12 визуально тяжело определить.
Значения NaN в locality_name, cityCenters_nearest, days_exposition,  Возможные потери данных.

Неслучайные ошибки:
Значение 0 в total_images. Возможно кто-то специально не хочет показывать квартиру ввиду ее отсутсвия или желания показать квартиру только при встрече.
Значение времени 00:00:00 в first_day_exposition. Точное время отправки на сайт продажи не требуется, нужен только день.
Значения NaN в is_apartment. Около 21000 квартир не имеют статуса в этой категории. Возможно продавцы не указали, жилая квартира для постоянного проживания или апартаменты, апартаменты - совсем новый тип жилых помещений. Так как в апартаментах невозможна (пока что) прописка могли и намеренно утаить эту информацию.
Значения NaN в столбцах parks_nearest и ponds_nearest, parks_around3000 и ponds_around3000. Если парк или пруд не в пешей доступности, то скорее всего его не укажут.

Неявные дубликаты в locality_name - наличие в языке несколько общеупотребительных обозначений небольших населенных пунктов
Аномалии в количестве балконов - возможна нестандартная перепланировка (разделение лоджии на две например) или визуальная ошибка. Аномалии в высоте потолков - неправильная запись данных либо изнальчно либо при копироании.

Неверный тип данных, float вместо int наверняка возник при копировании или выгрузке данных

### Предобработка данных

#### Работа с пропусками

```python
data.dtypes
#тип данных и запишем необработанные данные как old_data для оценки очистки
old_data = data
```

```python
data.isna().sum() #...подсчет пропусков в каждом столбце
```

```python
#заменим тип данных и где явно можно заполнить пропуски по методу бритвы Оккама
#не указан балкон, парк, пруд по близости скорее всего их нет, апартаменты новый тип жилья поэтому вероятнее False
data['last_price'] = data['last_price'].astype(int)
data['balcony'] = data['balcony'].fillna(0).astype(int)
data['parks_around3000'] = data['parks_around3000'].fillna(0).astype(int)
data['ponds_around3000'] = data['ponds_around3000'].fillna(0).astype(int)
data['is_apartment'] = data['is_apartment'].fillna('False').astype(bool)
data['locality_name'] = data['locality_name'].fillna('Неизвестно')
```

```python
#переведем столбец с датой в формат даты без времени
data['first_day_exposition'] = pd.to_datetime(data['first_day_exposition'], format = '%Y-%m-%d')
```

```python
#поработаем над дубликатами в 'locality_name'
sorted(data['locality_name'].unique())
```

```python
#источником неявных дубликатов в данных являются админитративные обозначения, добавленные к названиям
#избавимся от них с помощью списка административных обозначений и цикла
administr_dict = ['поселок ', 'городского типа ', 'городской ',  'деревня ', 
                'коттеджный ', 'село ', 'посёлок ', 'при железнодорожной станции ',
                'станции ', 'садоводческое некоммерческое товарищество ', 'садовое товарищество ']

for value in administr_dict:
    data['locality_name'] = data['locality_name'].str.replace(value, '')
```

```python
sorted(data['locality_name'].unique()) #проверим теперь окончательный список поселений
```

```python
#проверим корректонсть замен
data.dtypes
```

### Посчитайте и добавьте в таблицу новые столбцы

```python
#новый столбец с ценой квадратного метра, округленный и приведенный к типу int
data['price_per_square_meter'] = round(data['last_price'] / data['total_area'])
data['price_per_square_meter'] = data['price_per_square_meter'].astype('int')
#новый столбец с обозначением дня недели продажи
data['weekday_exposition'] = data['first_day_exposition'].dt.weekday
#новый столбец с обозначением  месяца продажи
data['month_exposition'] = data['first_day_exposition'].dt.month
#новый столбец с обозначением года продажи
data['year_exposition'] = data['first_day_exposition'].dt.year

#определим функцию категоризации по этажам,

def floor_categ(row):
    if row['floor'] == 1:
        return 'первый'
    elif row['floor'] == row['floors_total']:
        return 'последний'
    else:
        return 'другой'
#используя метод apply применимо к каждой строке датафрейма через axis = 1
data['floor_type'] = data.apply(floor_categ, axis=1)
#новый столбец с расстоянием до центра в км, пропуски игнорим
data['city_сenter_km'] = round(data['city_center_nearest'] / 1000)

display(data.head(10))
```
