## Подготовка данных к анализу.

### Приведём названия столбцов к нижнему регистру.

```python
data_games.columns = data_games.columns.str.lower()
```

### Преобразуем данные в нужные типы.

- Столбец 'year_of_release':
    Преобразуем тип данных столбца year_of_release к целочисленному (int), т.к. год не может быть дробным (float).
- Столбец 'user_score'. Тип столбца 'User_Score' object, хотя оценка пользователей даётся в значениях типа float. Это связано с тем, что в данном столбце присутствует аббревиатура 'tbd' (to be defined - подлежащий определению). Указывается тогда, когда игра имеет недостаточное количество отзывов (менее 4) для подсчета рейтинга пользователей и его публикации. 
Тем не менее, числовой тип столбца необходимо задать, поскольку 'tbd' можно считать текстовым аналогом NaN, то, соответственно и заменим tbd и NaN, а затем преобразуем тип данных на float64.

```python
data_games['year_of_release'] = data_games['year_of_release'].astype('Int64')
```

```python
data_games['user_score'] = data_games['user_score'].replace('tbd', np.NaN)
data_games['user_score'] = data_games['user_score'].astype('float')
```

### Обработаем пропуски.

Пропуски в столбце 'name'.
При подсчёте общих пропусков - 'name' было всего 2, можно избавиться

```python
data_games = data_games.dropna(subset=['name'])
```

Пропуски в столбце 'year_of_release'.
Количесвто пропусков мало - 269 штук и составляет всего 1,6 %. От них можно избавиться.

```python
data_games = data_games.dropna(subset=['year_of_release'])
```

Пропуски в столбцах 'critic_score', 'user_score', 'rating'.
В этих столбцах слишком много пропусков. Удалять нельзя, заменять на средние или медианные значения - сильно сместить выборку, поэтому можно поставить например, значение 'Unknown', но при обработке данных его исключать и заменять тип данных. 
Данные пропуски вероятно связаны с тем, что оценки и рейтинги данным играм не выставлялись.

```python
data_games['critic_score'] = data_games['critic_score'].fillna('Unknown')
data_games['user_score'] = data_games['user_score'].fillna('Unknown')
data_games['rating'] = data_games['rating'].fillna('Unknown')
```

Исследуем, как менялись продажи по платформам. Выберем платформы с наибольшими суммарными продажами и построим распределение по годам. Рассмотрим, за какой характерный срок появляются новые и исчезают старые платформы.

**Количество продаж по всем платформам.**

```python
data_games.pivot_table(index='platform', values='total_sales', aggfunc='sum')\
.sort_values('total_sales', ascending=False)\
.plot(kind='bar', title='Количество продаж по платформам', figsize=(15, 10))

plt.xlabel('Платформа')
plt.ylabel('Суммарные продажи')
plt.show()
```

Наибольшее количество суммарных продаж на платформах: PS2, X360, PS3, Wii, DS, PS. Построим распределение по годам для данных платформ.

```python
#для этого напишем функцию для построения гистограмм продаж платформ, в качестве переменной возьмём не только название 
#платформы но и срезы датасета с которыми нам предстоит работать 
def hist_of_games(dataset, platform):
    dataset[dataset['platform'] == platform]\
        .pivot_table(index='year_of_release', values='total_sales', aggfunc='sum')\
        .plot(kind='bar', figsize=(6,8))
    plt.xlabel('Год выпуска')
    plt.ylabel('Суммарные продажи (млн.)')
    plt.title('Количество продаж ' + platform)
    plt.legend(['Общая выручка'])
```

```python
#здесь строим гистограммы для платформ по всем годам выпуска (далее посмотрим, что случилось с продажами после 2012го)
hist_of_games(data_games, 'PS2')
```

```python
hist_of_games(data_games, 'X360')
```

```python
hist_of_games(data_games, 'Wii')
```

```python
hist_of_games(data_games, 'DS')
```

## Портрет пользователя каждого региона.
Составим портрет пользователя каждого региона. 

Для этого определим для пользователя каждого региона (NA, EU, JP):
- самые популярные платформы (топ-5);
- самые популярные жанры (топ-5);
- влияние рейтинга ESRB на продажи в отдельном регионе.

```python
#Для выполнения всех этих задач напишем функцию, в качестве dataset - возьмём данные за последние три года (или другие)
#index - платформы, жанр, рейтинг; region - 'total-sales', 'na_sales', 'eu_sales', 'jp_sales'
#тип диаграммы - pie, т.к. исследуем доли определённых рынков
def pie_of_games(dataset, index, region):
    dataset\
        .pivot_table(index=index, values=region, aggfunc='sum')\
        .sort_values(region, ascending=False)\
        .plot(kind='pie', subplots=True, title=('Sales by ' + index + ' (' + region + ')'), figsize=(10, 10));
    print ('Top-5 in', region)
    display(dataset\
    .groupby([index])[region]\
    .agg(['sum']).sort_values(by='sum', ascending=False).head())
```

### Определим самые популярные платформы (топ-5).
**Посмотрим на общее количество региональных продаж по всем игровым платформам и выведем топ-5.**
- na_sales: продажи в Северной Америке.

```python
pie_of_games(data_games_2014_2016, 'platform', 'na_sales');    
```

- eu_sales: продажи в Европе.

```python
pie_of_games(data_games_2014_2016, 'platform', 'eu_sales');
```

- jp_sales: продажи в Японии.

```python
pie_of_games(data_games_2014_2016, 'platform', 'jp_sales');
```

## Проверка гипотез.

Предлагается проверить следующие гипотезы:
- Средние пользовательские рейтинги платформ Xbox One и PC одинаковые.
- Средние пользовательские рейтинги жанров Action и Sports разные.

Поскольку пользовательские рейтинги платформ Xone PC не зависят друг от друга то применяем стандартный t-тест о равенстве среднего двух генеральных совокупностей.

```python
data_games_PC_scored = data_games.query('platform == "PC" and user_score != "Unknown"').copy()
data_games_PC_scored['user_score'] = data_games_PC_scored['user_score'].astype('float64')

data_games_Xone = data_games.query('platform == "XOne" and user_score != "Unknown"').copy()
data_games_Xone_scored['user_score'] = data_games_Xone_scored['user_score'].astype('float64')

#сравним полученный массив оценок пользователей PC, с полученным ранее массивом оценок пользователей XOne
#data_games_Xone_scored['user_score']
```

```python
data_games_action_scored = data_games.query('genre == "Action" and user_score != "Unknown"').copy()
data_games_action_scored['user_score'] = data_games_action_scored['user_score'].astype('float64')

data_games_sports = data_games.query('genre == "Sports" and user_score != "Unknown"').copy()
data_games_sports_scored['user_score'] = data_games_sports_scored['user_score'].astype('float64')
```

Уровень достоверности **alpha** примем равным 5%.

### Гипотеза - Средние пользовательские рейтинги платформ Xbox One и PC одинаковые.
- Н0: Средние пользовательские рейтинги платформ Xbox One и PC одинаковые.
- Н1: Средние пользовательские рейтинги платформ Xbox One и PC разные.

```python
alpha = 0.05  
results = st.ttest_ind(data_games_Xone_scored['user_score'], data_games_PC_scored['user_score']) 

print('p-значение:', results.pvalue)

if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу')
```

**Вывод:** Полученное значение p-value больше заданного нами порогового значение alpha, вследствии чего, мы можем отвергнуть нулевую гипотезу, и, следовательно принять, что **средние пользовательские рейтинги платформ Xbox One и PC разные**.
